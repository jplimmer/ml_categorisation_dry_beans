## Data inspection

```python
df = pd.read_excel(path+'Dry_Bean_Dataset.xlsx')
df
```

ShapeFactor1:
$SF1 = \frac{L}{A}$

ShapeFactor2:
$SF2 = \frac{l}{A}$
 
ShapeFactor3:
$SF3 = \frac{A}{L2∗L2∗π}$
(0-1, how closely a shape resembles a perfect circle)

ShapeFactor4:
$SF4 = \frac{A}{L2∗l2∗π}$
(0-1, how closely a shape resembles a perfect ellipse)

## Train-test split

```python
X = df.drop("Class", axis=1)
y = df["Class"]

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
X_train.shape
```

```python
# Distribution of data by feature

for col in X.columns:
    sns.histplot(data=df, x=col)
    plt.show()
```

## PCA

```python
shades = ["#5961A6", "#A65988", "#59A677", "#A69E59"]
```

```python
# Are some of the measurements much more significant?

scaler = StandardScaler()
decomp = PCA()
pipe = make_pipeline(scaler, decomp)
pipe.fit(X_train, y_train)

explained_variance = pipe['pca'].explained_variance_ratio_.cumsum()
index = [i+1 for i in range(len(explained_variance))]

pca_fig, ax = plt.subplots(figsize=(10, 6))

sns.lineplot(x=index, y=explained_variance, color=shades[0])
sns.scatterplot(x=index, y=explained_variance, s=100, color=shades[0])

ax.hlines(y=0.95, xmin=1, xmax=len(explained_variance), color='#8b0000', linestyle='--')
ax.vlines(x=4, ymin=0, ymax=0.95, color=shades[0], alpha=0.5, linestyle=":")
xs = ax.lines[0].get_xdata()
ys = ax.lines[0].get_ydata()
ax.fill_between(xs, 0, ys, where=(xs<=4), interpolate=True, color=shades[0], alpha=0.2)

plt.xlim((1-0.2, len(explained_variance)+0.2))
plt.ylim((0, 1.1))
x_s, x_e = ax.get_xlim()
ax.xaxis.set_ticks(np.arange(x_s+0.2, x_e))

sns.despine()
plt.ylabel('Total Explained Variance Ratio', fontsize=15)
plt.xlabel('Number of Principal Components', fontsize=15)
plt.title("Explained Variance per Additional PC", fontsize=18)
plt.show()
pca_fig.savefig('beans_pca.png', transparent=True, bbox_inches="tight")
```

```python
scaler = StandardScaler()
decomp = PCA(n_components=0.95)
pipe = make_pipeline(scaler, decomp)
pipe.fit(X_train, y_train)

loadings = pipe['pca'].components_

# Iterate over PCs to get top contributing features to each PC 
for i, component in enumerate(loadings):
    sorted_comp_index = np.argsort(np.abs(component))[::-1]
    
    # Print results
    print(f"Principal Component {i+1}")
    for feature_idx in sorted_comp_index:
        print(f"{X.columns[feature_idx]}: {component[feature_idx]}")
        
    print()
```

```python
# PC1 - 'Length' (mostly from MajorAxisLength (small SF2, perimeter, equivdiameter, ConvexArea, Area))
#     - smaller SF2 = lower ratio of width to area

# PC2 - 'Width' (MinorAxisLength)
#     - lower SF1 = lower ratio of length to area
#     - lower AspectRatio = lower ratio of length to width
#     - Compactness = another version of roundness

# PC3 - 'Concavity'
#     - lower Solidity (how well area corresponds to smallest convex polygon (i.e. limited concave bends))
#     - lower SF4 = doesn't look like a perfect ellipse

# PC3 - 'Squareness'
#     - Extent = fits very well within bounding box - rectangularish?
```

## Classifier evaluation

```python
decomps = [None, PCA(n_components=0.95)]

classifiers = [LogisticRegression(max_iter=1000),
               DecisionTreeClassifier(),
               RandomForestClassifier(n_jobs=-1), 
               SVC(cache_size=1000),
               KNeighborsClassifier(n_jobs=-1)]
```

```python
for d in decomps:
    print(d, "\n")
    for clf in classifiers:
        scaler = StandardScaler()
        pipe = make_pipeline(scaler, d, clf)
        grid = GridSearchCV(pipe, param_grid={}, cv=5, scoring="accuracy")
        grid.fit(X_train, y_train)
        print(f"{clf} best fold score: {grid.best_score_}")
        print(f"\t(fold score std: {grid.cv_results_['std_test_score']})\n")
    print("", "")
```

## Hyper-parameter tuning

```python
for clf in classifiers:
    print(type(clf).__name__.lower()+"__")
```

### `LogReg`

```python
param_grid_lr = {
    'logisticregression__C':[0.001, 0.01, 0.1, 1, 1.5, 2, 5],
}

pipe_lr = make_pipeline(StandardScaler(), LogisticRegression(max_iter=1000))
grid_lr = GridSearchCV(pipe_lr, param_grid_lr, cv=5, scoring="accuracy")
grid_lr.fit(X_train, y_train)
grid_lr_pred = grid_lr.predict(X_train)

print(f"Best mean CV score: {grid_lr.best_score_}")
print(f"Best params: {grid_lr.best_params_}\n")

print(f"Accuracy (for whole training set): {accuracy_score(y_train, grid_lr_pred)}")
ConfusionMatrixDisplay.from_predictions(y_train, grid_lr_pred)
```

```python
ConfusionMatrixDisplay.from_predictions(y_train, grid_lr_pred, normalize="true")
```

### `SVC`

```python
param_grid_svc = {
    'svc__C':[0.1, 1, 2, 5],
    'svc__kernel':['linear', 'poly', 'rbf', 'sigmoid'],
    'svc__gamma':['scale', 1]
}

pipe_svc = make_pipeline(StandardScaler(), SVC(cache_size=1000))
grid_svc = GridSearchCV(pipe_svc, param_grid_svc, cv=5, scoring="accuracy")
grid_svc.fit(X_train, y_train)
grid_svc_pred = grid_svc.predict(X_train)

print(f"Best mean CV score: {grid_svc.best_score_}")
print(f"Best params: {grid_svc.best_params_}\n")

print(f"Accuracy (for whole training set): {accuracy_score(y_train, grid_svc_pred)}")
ConfusionMatrixDisplay.from_predictions(y_train, grid_svc_pred)
```

### `RandomForestClassifier`

```python
param_grid_rf = {
    'randomforestclassifier__n_estimators':[50, 100, 150, 200],
    'randomforestclassifier__min_samples_split':[2, 3, 4],
    'randomforestclassifier__max_depth':[None, 10, 50],
    'randomforestclassifier__max_features':[2, 3]
}

pipe_rf = make_pipeline(StandardScaler(), RandomForestClassifier(n_jobs=-1))
grid_rf = GridSearchCV(pipe_rf, param_grid_rf, cv=5, scoring="accuracy")
grid_rf.fit(X_train, y_train)
grid_rf_pred = grid_rf.predict(X_train)

print(f"Best mean CV score: {grid_rf.best_score_}")
print(f"Best params: {grid_rf.best_params_}\n")

print(f"Accuracy (for whole training set): {accuracy_score(y_train, grid_rf_pred)}")
ConfusionMatrixDisplay.from_predictions(y_train, grid_rf_pred)
```
